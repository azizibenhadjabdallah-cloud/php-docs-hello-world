```python
import pandas as pd

customers = pd.read_csv("/content/sample_data/customer.csv")
products = pd.read_csv("/content/sample_data/products.csv")
orders = pd.read_csv("/content/sample_data/orders.csv")
order_items = pd.read_csv("/content/sample_data/orderitems.csv")
```

```python
products[products["stock"] < 0]
```

```python
products[products["unit_price"] <= 0]
order_items[order_items["unit_price"] <= 0]
```

```python
order_items[order_items["quantity"] <= 0]
```

```python
valid_categories = ["Laptop","Smartphone","Tablet","TV","Accessory",
                    "Headphones","Console","Smartwatch","Camera"]
products[~products["category"].isin(valid_categories)]
```

```python
orders["order_date"] = pd.to_datetime(orders["order_date"])
orders[orders["order_date"] > "2025-01-01"]
```

```python
set_customers = set(customers["customer_id"])
orders[~orders["customer_id"].isin(set_customers)]
```

```python
set_products = set(products["product_id"])
order_items[~order_items["product_id"].isin(set_products)]
```

```python
products.loc[products["stock"] < 0, "stock"] = 0
```

```python
valid_price_mask = products["unit_price"] > 0
median_price = products.loc[valid_price_mask, "unit_price"].median()
```

```python
products.loc[~valid_price_mask, "unit_price"] = median_price
```

```python
products.loc[~products["category"].isin(valid_categories), "category"] = "Accessory"
products["category"] = products["category"].fillna("Accessory")
```

```python
orders_clean = orders[orders["order_total"] > 0].copy()
```

```python
orders_clean["order_date"] = pd.to_datetime(orders_clean["order_date"], errors="coerce")

max_allowed_date = pd.Timestamp("2025-01-01")
```

```python
future_mask = orders_clean["order_date"] > max_allowed_date
orders_clean.loc[future_mask, "order_date"] = max_allowed_date
```

```python
order_items_clean = order_items[order_items["quantity"] > 0].copy()
```

```python
valid_products = set(products["product_id"])
order_items_clean = order_items_clean[order_items_clean["product_id"].isin(valid_products)]
```

```python
valid_orders = set(orders_clean["order_id"])
order_items_clean = order_items_clean[order_items_clean["order_id"].isin(valid_orders)]
```

```python
order_items_clean["line_total"] = order_items_clean["quantity"] * order_items_clean["unit_price"]
```

```python
totals = (order_items_clean
          .groupby("order_id")["line_total"]
          .sum()
          .reset_index()
          .rename(columns={"line_total":"order_total_calc"}))

orders_clean = orders_clean.merge(totals, on="order_id", how="left")
```

```python
orders_clean["order_total"] = orders_clean["order_total_calc"].fillna(orders_clean["order_total"])
orders_clean = orders_clean.drop(columns=["order_total_calc"])
```

```python
products.to_csv("products_clean.csv", index=False)
orders_clean.to_csv("orders_clean.csv", index=False)
order_items_clean.to_csv("order_items_clean.csv", index=False)
```

```python
customers["gender"] = customers["gender"].str.strip()

valid_gender = ["Male", "Female"]
customers.loc[~customers["gender"].isin(valid_gender), "gender"] = "Unknown"
customers["gender"] = customers["gender"].fillna("Unknown")
```

```python
customers["birthdate"] = pd.to_datetime(customers["birthdate"], errors="coerce")


customers = customers[
    (customers["birthdate"] >= "1940-01-01") &
    (customers["birthdate"] <= "2010-12-31")
]
```

```python
customers_clean = customers.copy()
customers_clean = customers_clean.drop(columns=["phone"])
customers_clean = customers_clean.rename(columns={"phone_clean": "phone"})

customers_clean.to_csv("customers_clean.csv", index=False)
```

```python
dim_customer = customers.copy()


dim_customer.insert(0, "customer_key", range(1, len(dim_customer) + 1))

dim_customer = dim_customer[["customer_key", "customer_id", "gender", "birthdate", "email"]]

dim_customer.to_csv("DimCustomer.csv", index=False)
```

```python
dim_product = products.copy()
dim_product.insert(0, "product_key", range(1, len(dim_product) + 1))

dim_product = dim_product[[
    "product_key", "product_id", "category", "brand", "product_name", "unit_price"
]]

dim_product.to_csv("DimProduct.csv", index=False)
```

```python
orders["order_date"] = pd.to_datetime(orders["order_date"], errors="coerce")

unique_dates = orders["order_date"].dropna().drop_duplicates().sort_values()

dim_date = pd.DataFrame({"date": unique_dates})
dim_date.insert(0, "date_key", range(1, len(dim_date) + 1))
dim_date["year"] = dim_date["date"].dt.year
dim_date["month"] = dim_date["date"].dt.month
dim_date["day"] = dim_date["date"].dt.day

dim_date.to_csv("DimDate.csv", index=False)
```
